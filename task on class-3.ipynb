```python
"""
Write a Python program that takes the user’s full name as input and then prints the datatype of the input,
the first character of the name using indexing, the last four characters using slicing, the full name reversed using a negative slicing step,
and finally the total number of characters in the name.
"""
```

```python
Name=input("Enter your name:")
print(f"first character of the name:{Name[0]}")
print(f"length of the name {len(Name)}")
print(f"last four charaters of the name :{Name[-4:]}")
print(f"reverse of name: {Name[::-1]}")
print(f"length of the Name:{len(Name)}")
```
