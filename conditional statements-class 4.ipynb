```python
# if statement
age=int(input("Enter your age"))
if age>=18:
    print("you can vote!!!!")
```

```python
# if else statement
age=int(input("enter your age:"))
if age>=18:
    print("you can vote")
else:
    print("you can't vote")
```

```python
#Multiple conditions check
marks=int(input("enter your marks:"))
if marks>=90:
    print("you got A grade ")
elif marks>=80:
    print("you got B grade")
elif marks>=70:
    print("you got C grade")
elif marks>=60:
    print("you got D grade")
elif marks>=50:
    print("you got E garde")
else:
    print("you got F grade !!!!!!!")
```

```python
# Nested if
marks=int(input("Enter your marks:"))
Attendance=int(input("Enter your attendance % :"))
if marks>=85:
    if Attendance>=80:
        print("You are eligible for scholarship")
    else:
        print("you have less attendance")
else:
    print("Marks too low")
```

```python
#Multiple Conditions (Logical Operators)
#and
age=int(input("Enter your age:"))
income=int(input("Enter your income :"))
if age>=23 and income>=25000:
    print("You are eligible for credit card")
else:
    print("Sorry You are not eligible")
```

```python
#or
day=input("enter a day:")
if day=="sunday" or day=="saturday":
    print("weekend")
else:
    print("weekday")
```

```python
#not
sleep=input("are you feeling sleepy (yes/no):")
if sleep != "yes":
    print("watch TV")
else:
    print("go and sleep")
```
