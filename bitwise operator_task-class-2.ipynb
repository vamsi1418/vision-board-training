```python
#bitwise operator task....
a=5                 #  0 1 0 1 = 5  
b=3                 #  0 0 1 1 = 3
#AND                   0 0 0 1 = 1                            
print(f" value is {a&b}")
```

```python
#OR                                        0 1 0 1 = 5
#                                          0 0 1 1 = 3
#                                          0 1 1 1 = 7      
print(f" value is {a|b}")
```

```python
#XOR                                       0 1 0 1 = 5
#                                          0 0 1 1 = 3
#                                          0 1 1 0 = 6
print(f" value is {a^b}")
```

```python
#NOT
print(f"value a {~a}")            # 0 1 0 1 = 5
                                  # 1 0 1 0 =-6
print(f"value b {~b}")             #  0 0 1 1 = 3
                                    # 1 1 0 0 = -4
```

```python
#rightshift 
print(f" value a {a>>2}")         # 0 1 0 1 = 5
                                   # 0 0 0 1 = 1
print (f" value b {b>>2}")         # 0 0 1 1 = 3
                                    # 0 0 0 0= 0
```

```python
#leftshift
print(f" value a {a<<1 }")           # 0 1 0 1 = 5
                                    #  1 0 1 0  =10
print (f" value b {b<<2}")          # 0 0 1 1 = 3
                                    # 1 1 0 0 = 12
```
